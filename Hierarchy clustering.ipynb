```python
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
import numpy as np
from pandas import DataFrame
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
import pandas as pd
from scipy.stats import zscore
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
df=pd.read_csv('mtcars.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
Z = linkage(df,'single')
dendrogram(Z)
```

```python
plt.figure(figsize=(25,15))
plt.title('Heirachial clustering dendrogram- Single Linkage')
plt.xlabel('sample index')
plt.xlabel('distance')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=20.)
plt.show()
```

```python
Z1 = linkage(df,'average')
dendrogram(Z1)
```

```python
plt.figure(figsize=(25,15))
plt.title('Heirachial clustering dendrogram- Average Linkage')
plt.xlabel('sample index')
plt.xlabel('distance')
dendrogram(Z1,leaf_rotation=90.,leaf_font_size=20.)
plt.show()
```

```python
Z2 = linkage(df,'centroid')
dendrogram(Z2)
```

```python
plt.figure(figsize=(25,15))
plt.title('Heirachial clustering dendrogram- Centroid Linkage')
plt.xlabel('sample index')
plt.xlabel('distance')
dendrogram(Z2,leaf_rotation=90.,leaf_font_size=20.)
plt.show()
```
